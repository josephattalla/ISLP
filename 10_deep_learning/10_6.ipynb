### 6. Consider the simple function R(β) = sin(β) + β/10.

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### (a) Draw a graph of this function over the range β ∈ [−6, 6].

```python
def R(beta):
    return np.sin(beta) + beta/10
```

```python
beta = np.linspace(-6, 6, 1000)
plt.plot(beta, R(beta))
```

##### (c) Given β0 = 2.3, run gradient descent to fnd a local minimum of R(β) using a learning rate of ρ = 0.1. Show each of β0, β1,... in your plot, as well as the final answer.

```python
def dR_dbeta(beta):
    return np.cos(beta) + 1/10


def gradient_descent(beta_initial, learning_rate, threshold, max_iterations):
    beta_values = [beta_initial]
    deriv = dR_dbeta(beta_initial)
    iteration = 0
    while abs(deriv) > threshold and iteration < max_iterations:
        beta_initial = beta_initial - learning_rate*deriv
        beta_values.append(beta_initial)
        deriv = dR_dbeta(beta_initial)
        iteration += 1
    return beta_values
```

```python
beta_values = gradient_descent(2.3, 0.1, 0.0001, 100000000)

beta_values = np.asarray(beta_values)
R_values = R(beta_values)

plt.figure(figsize=(8, 6))
plt.plot(beta_values, R_values, '-o', label=r'$R(\beta) = \sin(\beta) + \frac{\beta}{10}$')
plt.title('Gradient Descent to Find Local Minimum')
plt.xlabel('β')
plt.ylabel('R(β)')
plt.grid(True)
plt.legend()
plt.show()
```

##### (d) Repeat with β0 = 1.4.

```python
beta_values = gradient_descent(1.4, 0.1, 1e-10, 1e10)
beta_values = np.asarray(beta_values)
R_values = R(beta_values)
plt.plot(beta_values, R_values, marker='o')
plt.grid()
```
